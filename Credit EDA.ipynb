### Imputation Reporting

- AMT_ANUITY has missing values. It is series of identical installments to be paid by the client and it is calculated by using annual interest rate. Since, the company has the knowledge of Annual Interest Rate it is charging and also the AMT_CREDIT value is known, the AMT_ANNUITY can be calculated by the Company itself. 

- AMT_GOODS_PRICE has missing at random values.It can be imputed with the value of AMT_CREDIT. Generally, AMT_GOODS_PRICE is equal to AMT_CREDIT unless the client has insurance amount. 

- NAME_TYPE_SUIT has missing completely at random values. It is not dependent on any feature of dataset. However, some relation can be found between NAME_TYPE_SUIT and NAME_FAMILY_STATUS, the missing values can be filled with Unaccompanied FOR Single, Not married, Widow, Separated. Otherwise, for Married clients it can be filled with Family. 

- EXT_SOURCE_2 and EXT_SOURCE_3 also has missing values. But, EXT_SOURCE_3 has more missing value percentage than EXT_SOURCE_2. So, if the values rows(clients) are common for missing values of both sources, then it is Missing not at Random values. Maybe the client never opted for loan. Hence, these values can be filled with NA. 

- In EXT_SOURCE_3, the missing values are at random. Hence, it should be filled with the values of EXT_SOURCE_2. 
  i.e., for a client having missing value in source 3 can be given the same score as source 2.
